**Task 1**

Computer VS Player

Player Guessing the Number given by computer






```python
import random

def non_ai_game():
    target = random.randint(1, 100)
    print("Welcome to the Number Guessing Game!")
    print("I'm thinking of a number between 1 and 100. Try to guess it!")

    while True:
        try:
            guess = int(input("Enter your guess: "))
            if guess < target:
                print("Too low! Try again.")
            elif guess > target:
                print("Too high! Try again.")
            else:
                print(f"Congratulations! You guessed the number {target}.")
                break
        except ValueError:
            print("Please enter a valid number.")

non_ai_game()
```

**Task 2**

AI Binary Search

AI guessing the number selected by user

```python
import random

def binary_search_game(target, low=1, high=100):
    guess_count = 0
    while low <= high:
        guess_count += 1
        mid = (low + high) // 2
        print(f"Computer guesses: {mid}")

        if mid == target:
            print(f"Computer found the number {target} in {guess_count} guesses!")
            return
        elif mid < target:
            low = mid + 1
        else:
            high = mid - 1

    print("Computer could not find the number.")


```

**Task # 3**

AI BFS
AI guesses a number selected by the user using breadth-first search. The primary function of breadth-first search is to explore all the options at the current level before moving deeper. In this context, BFS will behave similarly to binary search, as it halves the range and guesses the middle element each time. The key difference from binary search is how BFS stores and processes the guesses: instead of using a single variable, BFS uses a queue, ensuring that guesses are processed in a first-in, first-out (FIFO) manner, exploring all possibilities at each level before proceeding.

This has no benefits over a binary search in this context.

```python
from collections import deque

def bfs_guess_game(target, low=1, high=100):
    guess_count = 0
    queue = deque(range(low, high + 1))

    while queue:
        guess = queue.popleft()
        guess_count += 1
        print(f"Computer guesses: {guess}")

        if guess == target:
            print(f"Computer found the number {target} in {guess_count} guesses!")
            return

    print("Computer could not find the number.")

# Example usage
target_number = random.randint(1, 100)  # Set this to the number the player is thinking of
bfs_guess_game(target_number)
```

**Task 4**

AI DFS



```python
def dfs_guess_game(target, low=1, high=100):
    guess_count = 0
    stack = list(range(low, high + 1))

    while stack:
        guess = stack.pop()
        guess_count += 1
        print(f"Computer guesses: {guess}")

        if guess == target:
            print(f"Computer found the number {target} in {guess_count} guesses!")
            return

    print("Computer could not find the number.")

# Example usage
target_number = random.randint(1, 100)  # Set this to the number the player is thinking of
dfs_guess_game(target_number)
```

**Task 5**

Simulated Annealing Version

Simulated Annealing is a probabilistic technique for approximating the global optimum. It allows occasional jumps to worse guesses to avoid local minima.


```python
import random
import math

def simulated_annealing_game(target, low=1, high=100, initial_temp=100, cooling_rate=0.95):
    guess_count = 0
    current_guess = random.randint(low, high)
    best_guess = current_guess
    temperature = initial_temp

    while temperature > 1:
        guess_count += 1
        print(f"Computer guesses: {current_guess}")

        if current_guess == target:
            print(f"Computer found the number {target} in {guess_count} guesses!")
            return

        # Simulate random jump
        new_guess = random.randint(low, high)
        delta = abs(target - new_guess) - abs(target - current_guess)

        # Decide whether to accept the new guess
        if delta < 0 or random.random() < math.exp(-delta / temperature):
            current_guess = new_guess
            if abs(target - current_guess) < abs(target - best_guess):
                best_guess = current_guess

        temperature *= cooling_rate

    print("Computer could not find the number.")

# Example usage
target_number = random.randint(1, 100)  # Set this to the number the player is thinking of
simulated_annealing_game(target_number)
```
